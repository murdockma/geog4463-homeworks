````python
# boolean indexing to select all rows with California as the state
cali_fires = fire_occurrences_1992_2018[
    fire_occurrences_1992_2018['STATE'] == 'CA']

# grouping the california df on year, and aggregating ID and size columns
cali_fire_measures = cali_fires.groupby("FIRE_YEAR").agg(
    mean_size_acres=('FIRE_SIZE', 'mean'),
    total_fires=('FOD_ID', 'count'))

cali_fire_measures
````
